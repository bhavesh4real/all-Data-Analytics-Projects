# Sea Level Predictor
### Objective:

The objective of the Sea Level Predictor project is to analyze a dataset on global average sea level change since 1880 and use it to make predictions about future sea level rise.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('data/epa-sea-level.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

[dpi-fig relationship](https://stackoverflow.com/questions/47633546/relationship-between-dpi-and-figure-size)

```python
fig = plt.figure(figsize=(8,4),dpi=100)
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],c='#00008b',alpha= 0.45)

plt.xlabel('Year')
plt.ylabel('Sea Level (in inches)')

plt.title('Rise in Sea Level')
```

```python
lin_out = linregress(df['Year'],df['CSIRO Adjusted Sea Level']) #using linregress

slope = lin_out[0]
intercept = lin_out[1]
rvalue = lin_out[2]
pvalue = lin_out[3]
std_err = lin_out[4]

print(f'slope = {slope}\nintercept = {intercept}\nrvalue = {rvalue}\npvalue = {pvalue}\n\
StandardError = {std_err}')
```

```python
x_fit = np.linspace(np.min(df['Year']),2050)
y_fit = (x_fit * slope) + intercept  #y=(slope*x)+intercept
plt.plot(x_fit,y_fit,'orange')
```

```python
df_2k = df[df['Year']>=2000]
```

```python
lin_out_2k = linregress(df_2k['Year'],df_2k['CSIRO Adjusted Sea Level'])
```

```python
lin_out_2k
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],c='#00bfff',alpha= 0.70)

x_fit = np.arange(np.min(df['Year']),2050,1)
y_fit = (x_fit * slope) + intercept    #(slope*x)+intercept
plt.plot(x_fit,y_fit,color ='#8b0000')  #color ='#00008b'

x2_fit = np.arange(np.min(df_2k['Year']),2050,1)
y2_fit = (x2_fit * lin_out_2k.slope) + lin_out_2k.intercept
plt.plot(x2_fit,y2_fit,color = '#0bda51')

plt.xlabel('Year')
plt.ylabel('Sea Level (in inches)')

plt.title('Rise in Sea Level')
plt.grid(True)
```
